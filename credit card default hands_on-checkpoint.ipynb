#There are 25 variables:

ID: ID of each client
LIMIT_BAL: Amount of given credit in NT dollars (includes individual and family/supplementary credit
SEX: Gender (1=male, 2=female)
EDUCATION: (1=graduate school, 2=university, 3=high school, 4=others, 5=unknown, 6=unknown)
MARRIAGE: Marital status (1=married, 2=single, 3=others)
AGE: Age in years
PAY_0: Repayment status in September, 2005 (-1=pay duly, 1=payment delay for one month, 2=payment delay for two months, â€¦ 8=payment delay for eight months, 9=payment delay for nine months and above)
PAY_2: Repayment status in August, 2005 (scale same as above)
PAY_3: Repayment status in July, 2005 (scale same as above)
PAY_4: Repayment status in June, 2005 (scale same as above)
PAY_5: Repayment status in May, 2005 (scale same as above)
PAY_6: Repayment status in April, 2005 (scale same as above)
BILL_AMT1: Amount of bill statement in September, 2005 (NT dollar)
BILL_AMT2: Amount of bill statement in August, 2005 (NT dollar)
BILL_AMT3: Amount of bill statement in July, 2005 (NT dollar)
BILL_AMT4: Amount of bill statement in June, 2005 (NT dollar)
BILL_AMT5: Amount of bill statement in May, 2005 (NT dollar)
BILL_AMT6: Amount of bill statement in April, 2005 (NT dollar)
PAY_AMT1: Amount of previous payment in September, 2005 (NT dollar)
PAY_AMT2: Amount of previous payment in August, 2005 (NT dollar)
PAY_AMT3: Amount of previous payment in July, 2005 (NT dollar)
PAY_AMT4: Amount of previous payment in June, 2005 (NT dollar)
PAY_AMT5: Amount of previous payment in May, 2005 (NT dollar)
PAY_AMT6: Amount of previous payment in April, 2005 (NT dollar)
default.payment.next.month: Default payment (1=yes, 0=no)

```python
# now lets build the model with the raw data itself
```

```python
y=df['is_defaulter']
x=df.drop('is_defaulter',axis=1)
```

```python
x.columns
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.80,random_state=100)                                  
```

```python
print(xtrain.shape)
print(ytrain.shape)
print(xtest.shape)
print(ytest.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DT=DecisionTreeClassifier()
```

```python
DT.fit(xtrain,ytrain)
```

```python
dt_y_pred=DT.predict(xtest)
dt_y_pred.shape
```

```python
from sklearn.metrics import roc_curve,roc_auc_score,classification_report,cohen_kappa_score,confusion_matrix,plot_confusion_matrix,accuracy_score
```

```python
print(classification_report(ytest,dt_y_pred))
```

```python
roc_auc_dt = roc_auc_score(ytest,dt_y_pred)
accuracy_dt = accuracy_score(ytest,dt_y_pred)
kappa_score_dt = cohen_kappa_score(ytest,dt_y_pred)  
print('roc_auc:',round(roc_auc_dt,2))
print('accuracy:',round(accuracy_dt,2))
print('kappa_score:',round(kappa_score_dt,2))
```

```python
#now we build a model  with random forest and check if the model improves 
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RF=RandomForestClassifier()
```

```python
RF.fit(xtrain,ytrain)
```

```python
rf_y_pred=RF.predict(xtest)
```

```python
rf_y_pred.shape
```

```python
print(classification_report(ytest,rf_y_pred))
```

```python
# we can see that the model performed better when compared to decesion tree
```

```python
roc_auc_rf = roc_auc_score(ytest,rf_y_pred)
accuracy_rf = accuracy_score(ytest,rf_y_pred)
kappa_score_rf = cohen_kappa_score(ytest,rf_y_pred)  
print('roc_auc:',round(roc_auc_rf,2))
print('accuracy:',round(accuracy_rf,2))
print('kappa_score:',round(kappa_score_rf,2))
```

```python
plot_confusion_matrix(RF,xtest,ytest)
plt.grid(False)
```
